#### CLEANING

```python
# set ship date and order date to datetime Dtype
superstore_data["Order Date"] = pd.to_datetime(superstore_data["Order Date"])
superstore_data["Ship Date"] = pd.to_datetime(superstore_data["Ship Date"])
superstore_data.info()
```

```python
#create new columns

#Order ship time in days
superstore_data["Order Ship Time - Days"] = superstore_data["Ship Date"] - superstore_data["Order Date"]
superstore_data["Order Ship Time - Days"] = superstore_data["Order Ship Time - Days"].dt.days

#seperate out the order dates into year, month, day 
superstore_data["Order Year"] = superstore_data["Order Date"].dt.year
superstore_data["Order Month"] = superstore_data["Order Date"].dt.strftime("%b")
superstore_data["Order Day"] = superstore_data["Order Date"].dt.strftime("%a")

#remove old column
#superstore_data = superstore_data.drop("Order Ship Time", axis=1)

superstore_data.describe()
```

### ANALYSIS

### Regional Breakdown by Sales and Profit

```python
region_sales = superstore_data.groupby("Region")["Sales"].mean()
region_profit = superstore_data.groupby("Region")["Profit"].mean()

region_sales.sort_values(ascending=False).plot(kind="bar", title="Avg Sales by Region", ylabel="Avg Sales ($)", color="blue")
plt.show()
region_profit.sort_values(ascending=False).plot(kind="bar", title="Avg Profit by Region", ylabel="Avg profit ($)", color="orange")
plt.show()
```

##### South Region has the highest avg sales, while West Region has the highest avg profit

### Profit across all regions by Segment, category, and sub-category

```python
print(superstore_data.groupby(["Segment"], sort=False)[["Profit"]].mean().round(decimals=2).sort_values("Profit", ascending=False))
print(" ")
print(superstore_data.groupby(["Category"], sort=False)[["Profit"]].mean().round(decimals=2).sort_values("Profit", ascending=False))
print(" ")
print(superstore_data.groupby(["Category","Sub-Category"])[["Profit"]].mean().round(decimals=2).sort_values(["Category","Profit"], ascending=False))
print(" ")
print(superstore_data.groupby(["Region","Category"], sort=False)[["Profit"]].mean().round(decimals=2).sort_values(["Region","Profit"], ascending=False))

```

##### Furiniture is least profitable overall
##### Supplies, Bookcases, Tables are least profitable sub-cats with Bookcases and Tables both in Furniture category
##### Central is only region with negative profit category

### Sales and profit in each region by year

```python
list_of_regions = ["West", "Central", "East", "South"]

for item in list_of_regions:
    filtered_df = superstore_data[superstore_data["Region"] == item]
    
    filtered_df.groupby(["Region","Order Year"])["Sales"].mean().plot(title="Sales in " + item + " Region", ylabel="Avg Sales ($)", color="blue")
    plt.show()
    filtered_df.groupby(["Region","Order Year"])["Profit"].mean().plot(title="Profit in " + item + " Region", ylabel="Avg Profit ($)", color="orange")
    plt.show()

```

##### West has overall sales growth but trending down, profit is trending upward
##### Central is seeing downward trend in sales, profit has overall growth but is trending down
##### East is seeing a steady drop in sales, but an overall growth in profit trending upward
##### South is seeing an overall drop in sales but trending up, while overall profit has dropped

### Drilling down into Central Region

```python
#Customer segments
central_region_df = superstore_data[superstore_data["Region"] == "Central"]
central_region_df.groupby("Segment")["Segment"].count().plot(kind="pie", title= "Central Region Customer Segments", ylabel="")
```

##### Consumer segment account for more than half of all orders

```python
#Most popular category in central region consumer segment

central_consumer_df = central_region_df[central_region_df["Segment"] == "Consumer"]
print(central_consumer_df.groupby(["Category"], sort=False)["Quantity"].sum().sort_values(ascending=False))
print(" ")
print(central_consumer_df.groupby(["Category"], sort=False)["Profit"].mean().sort_values(ascending=False))
```

##### Office supplies have sold more than double the other categories
##### Technology has the highest profit, office supplies profit is low, furniture is negative

```python
#sub-categories in furniture category of consumer segment in central region

central_con_cat = central_consumer_df[central_consumer_df["Category"] == "Furniture"]
print(central_con_cat.groupby(["Sub-Category"], sort=False)["Quantity"].sum().sort_values(ascending=False))
print(" ")
print(central_con_cat.groupby(["Sub-Category"], sort=False)[["Profit"]].mean().round(decimals=2).sort_values("Profit", ascending=False))
```

##### Furnishings & chairs most popular product sub-category in furniture and Bookcases are least popular sub-category
##### Chairs are the only profitable sub-category

### Calculate how divesting unprofitable categories changes Profit for Central Region

```python
#overall profit Central Region

central_profit = round(central_region_df["Profit"].mean(),2)

print(central_profit)
```

```python
#adjusted profit of Central Region without negative sub-cats in furniture

list_of_subs = ["Furnishings", "Bookcases", "Tables"]
adjusted_profit_df = central_region_df.loc[~central_region_df["Sub-Category"].isin(list_of_subs)]

adjusted_central_profit = round(adjusted_profit_df["Profit"].mean(),2)
print(adjusted_central_profit)
```

```python
#percent change Central

central_percent_change = ((adjusted_central_profit - central_profit) / central_profit) * 100

print(str(round(central_percent_change, 2)) + "%")
```

##### Divesting unprofitable products will improve the Central Region profitability by 44.12%

```python
#overall superstore profit

superstore_profit = round(superstore_data["Profit"].mean(),2)

print(superstore_profit)
```

```python
#overall superstore profit without negative sub-cats in furniture

adjusted_total_profit_df = superstore_data.loc[~superstore_data["Sub-Category"].isin(list_of_subs)]

adjusted_superstore_profit = round(adjusted_total_profit_df["Profit"].mean(),2)

print(adjusted_superstore_profit)
```

```python
#percent change overall

overall_percent_change = ((adjusted_superstore_profit - superstore_profit) / superstore_profit) * 100

print(str(round(overall_percent_change, 2)) + "%")
```

##### Divesting unprofitable products will improve the overall Superstore profitability by 21.04%
